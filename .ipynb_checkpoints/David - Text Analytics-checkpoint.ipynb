### Pre-processing

Involves several steps, most of which have been combined in the code sample below:
- Selection - picked the review field only for cleaning and analysis
- Cleaning - removed all special characters, includes expunging html tags
- Transformation - Text is split in separate word arrays
- Integration - performed when performing visualizations

##### remove all html and special characters

```python
## convert all to lower case
r_lower = df['review'].str.lower()

## remove all html tags
# r_l_cleaned = list(map(lambda s: re.sub('<[^<]+?>', ' ', s), r_lower))
r_l_cleaned = r_lower.str.replace('<[^<]+?>', ' ')

# remove punctuations
r_l_cleaned = r_l_cleaned.str.replace('[^A-Za-z0-9\s]+', '')

# txt_s = pd.Series(r_l_cleaned.str.sub())
splt_r = r_l_cleaned.str.split(' ')

display(len(splt_r))
display(splt_r[:1])
```

### Collocations
Used Bigrams

```python
# pip install networkx
# pip install nltk
```

```python
from nltk import bigrams
import itertools
import collections
import time
import networkx as nx
```

```python
# add space to remove all spaces
stopwords.add('')
```

```python
st = time.time()

clean_splt_r = [[word for word in review if not word in stopwords] for review in splt_r]

terms_bigram = [list(bigrams(txt)) for txt in clean_splt_r]

et = time.time()

print('elapsed: {:.2f} seconds'.format(et-st))
print(len(terms_bigram))
```

#### get most common bigrams

```python
# Flatten list of bigrams in clean tweets
bigrams = list(itertools.chain(*terms_bigram))

# Create counter of words in clean bigrams
bigram_counts = collections.Counter(bigrams)

bigram_counts.most_common(20)
```

```python
bigram_df = pd.DataFrame(bigram_counts.most_common(100),
                             columns=['bigram', 'count'])

bigram_df

# Create dictionary of bigrams and their counts
d = bigram_df.set_index('bigram').T.to_dict('records')

# Create network plot 
G = nx.Graph()

# Create connections between nodes
for k, v in d[0].items():
    G.add_edge(k[0], k[1], weight=(v * 10))
    
fig, ax = plt.subplots(figsize=(35, 35))

pos = nx.spring_layout(G, k=2)

# Plot networks
nx.draw_networkx(G, pos,
                 font_size=16,
                 width=3,
                 edge_color='grey',
                 node_color='purple',
                 with_labels = False,
                 ax=ax)

# Create offset labels
for key, value in pos.items():
    x, y = value[0]+.135, value[1]+.045
    ax.text(x, y,
            s=key,
            bbox=dict(facecolor='red', alpha=0.25),
            horizontalalignment='center', fontsize=13)
    
plt.show()
```

#### _Intepretation_

The network graph shows the bigram relations, where a word is represented by a node, and the relationship is represented via an edge (connecting line). More correlated words create a tight cluster between nodes. 
This means that after observing one of these words in the cluster, there's is a high likelyhood the closely correlated word(s) will also be observed

As per the provided corpus, the top 20 bigram tuples produced some interesting observations. One interesting observation is (new, york) tuple, meaning most reviews touched on location (either setting of the movie, or reviewers location), of New York which happens to be one of the biggest and famous cities in USA, and the world. Another observation of interest is ('high', 'school'), meaning most reviewers watched, or most reviews were on, high-school themed movies

('special', 'effects') was number two in the bigram most common tuples, meaning most reviewers are keen on the visuals of created graphics in movies.
